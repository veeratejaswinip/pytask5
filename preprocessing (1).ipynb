```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"D:\taskwordfiles\healthcare.csv")
grouped_data = df.groupby('Neighbourhood')[['Scholarship', 'SMS_received']].sum()
print(grouped_data.head())
grouped_data.plot(kind='bar', figsize=(12,6))
plt.xlabel('Neighbourhood')
plt.ylabel('Total Count')
plt.title('Scholarship & SMS Received Count by Neighbourhood')
plt.tight_layout()
plt.show()
```
